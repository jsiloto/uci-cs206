```python
import matplotlib.pyplot as plt

small = [ 128, 74.9]
ensemble = [ 211, 77.9]
large = [ 286, 83.3]

plt.figure(figsize=(9, 6), dpi=400)
plt.scatter(*small, label="Small",  s=200)
plt.scatter(*large, label="Large",  s=200)
plt.scatter(*ensemble, label="Ensemble",  s=200)
plt.xlabel("Best Latency (ms)")
plt.ylabel("Accuracy %")
plt.legend()



plt.savefig('scatter.pdf')  
```
